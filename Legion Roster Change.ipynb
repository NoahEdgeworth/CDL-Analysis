So the question still remains, why get rid of Standy over Purj? Lets go deeper into the individual stats per series for each player on the team. Maybe Purj has more upside or potential? Maybe he played better against the top tier teams? Lets find out.

```python
series_df = pd.read_csv('Series Stats/Player Game by Game_Series Results_Table.csv')
series_df
```

```python
average_kd = series_df['K/D'].mean()
```

```python
legion_df = series_df[series_df['Team']=='Legion']
legion_df
```

Lets compare Standy and Purj by series

```python
standy_purj_df = legion_df[(legion_df['Player']=='Standy') | (legion_df['Player']=='Purj')]
standy_purj_df.sort_values('Series ID', ascending=True)
```

```python
online_df = standy_purj_df[standy_purj_df['Date']< 'Jan 26, 2024']
online_df.sort_values('Series ID', ascending=True)
```

```python
standy_online_kd_avg = online_df[(online_df['Player']=='Standy')]['K/D'].mean()
standy_online_kd_avg
```

```python
standy_online_kd_max = online_df[(online_df['Player']=='Standy')]['K/D'].max()
standy_online_kd_max
```

```python
standy_online_kd_min = online_df[(online_df['Player']=='Standy')]['K/D'].min()
standy_online_kd_min
```

```python
purj_online_kd_avg = online_df[(online_df['Player']=='Purj')]['K/D'].mean()
purj_online_kd_avg
```

```python
purj_online_kd_max = online_df[(online_df['Player']=='Purj')]['K/D'].max()
purj_online_kd_max
```

```python
purj_online_kd_min = online_df[(online_df['Player']=='Purj')]['K/D'].min()
purj_online_kd_min
```

```python
pivot_online_df = online_df.pivot(index='Opponent', columns='Player', values='K/D')

ax = pivot_online_df.plot(kind='bar', figsize=(16,6))
ax.set_xlabel('Opponent')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

ax.set_ylabel('K/D Ratio')
ax.set_title('Standy vs Purj Online KD Comparison by Series')
plt.legend(title='Player')

plt.show()
```

Lets first take a look at online play. The graph above shows Purj(Blue) and Standy(Orange) and their KDs against their opponents in online matches. 

Purj started the season with a slightly better KD than Standy in matches vs Faze and Miami. Standy quickly heated up with a huge performance in their upset match vs Optic in qualifiers. Purj really struggled against some of the lower their teams, like the Royal Ravens, where he posted a .53 kd for the series.
 
Standy had a higher KD in 4/7 online matches for Major 1. Standy had an average KD for online of .92 and Purj held a .83. Standy had a high KD of 1.08 and Purj with a 1.0, both of which was against Optic. Finally, Standy had a low KD of .71 and Purj with a low KD of .53. 

Overall in terms of KD ratio, Standy outperformed Purj in all aspects of the game.
